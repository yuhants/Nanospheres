```python
numpoints = 8182 # Total number of points in waveform
freqlist = np.linspace(1, 151, 76)*1000 # frequencies for frequency comb
tottime = 100 # Longer totaltime for PSD
time = np.linspace(0, 0.001*tottime, numpoints*tottime) # make 8182 points equal to 0.001 second

totalwav = np.zeros(numpoints*tottime)
    # Product the waveform
for i in freqlist:
    totalwav = totalwav + np.sin(2*np.pi*i*time)
    # Make all numbers integer
signal = totalwav/np.max(totalwav)*512
#signal = signal.astype(int)
        # Create initialise fgen if it was not supplied
plt.plot(time[:8183], signal[:8183])
plt.show()

import scipy.signal as scisig

PSD = scisig.welch(signal, fs = 1000*8182, nperseg = 100*8182)

lb = 0
ub = 20000
plt.plot(PSD[0][lb:ub], PSD[1][lb:ub])
plt.yscale('log')
plt.show()
```
