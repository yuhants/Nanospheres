```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import csv
```

```python

data = []
filename1 = 'C:/Users/yuhan/nanospheres/Data/Monitoring/240327/No PI Loop Figs/temp.csv'
with open(filename1, newline='') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=';')
    for row in spamreader:
        data.append(row)

time = [float(i) for i in np.transpose(np.array(data[12:]))[0]]
T = [float(i) for i in np.transpose(np.array(data[12:]))[3]]
T1 = [float(i) for i in np.transpose(np.array(data[12:]))[5]]
H = [float(i) for i in np.transpose(np.array(data[12:]))[4]]

time = np.array(time)
T = np.array(T)
T1 = np.array(T1)
H = np.array(H)

filename2  = 'C:/Users/yuhan/nanospheres/Data/Monitoring/240327/No PI Loop Figs/analysed.mat'
data = sio.loadmat(filename2)

lb = 34500
ub = 38500
PSDy_max = []
for i in data['y']:
    maxy = np.argmax(i[lb:ub])+lb
    PSDy_max.append(maxy)

lb = 26500
ub = 29500
PSDx_max = []
for i in data['x']:
    maxx = np.argmax(i[lb:ub])+lb
    PSDx_max.append(maxx)

lb = 5600
ub = 8200
PSDz_max = []
for i in data['z']:
    maxz = np.argmax(i[lb:ub])+lb
    PSDz_max.append(maxz)

PSDy_max = np.array(PSDy_max)*10
PSDx_max = np.array(PSDx_max)*10
PSDz_max = np.array(PSDz_max)*10
LP = np.array(data['LP'][0])
time2 = np.linspace(0, 20*len(LP), len(LP))
time2 = time2*time[-1]/time2[-1]
```

```python
plt.plot(time2, PSDy_max)
plt.figure()
plt.plot(time2, PSDx_max)
plt.figure()
plt.plot(time2, PSDz_max)
```

```python
plt.plot(time/3600, T)
plt.plot(time/3600, T1)
plt.ylabel('Temperature (C)')
plt.xlabel('Time (hrs)')
plt.figure()
plt.plot(time/3600, H)
plt.ylabel('Rel Humidity (%)')
plt.xlabel('Time (hrs)')
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('time (s)')
ax1.set_ylabel('Percentage change (%)')
ax1.tick_params(axis='y')

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel('Temperature')  # we already handled the x-label with ax1
ax2.tick_params(axis='y')

fig.tight_layout()  # otherwise the right y-label is slightly clipped

ax1.plot(time2, PSDy_max/PSDy_max[0]-1, label = 'y central frequency')
ax1.plot(time2, PSDx_max/PSDx_max[0]-1, label = 'x central frequency')
ax1.plot(time2, PSDz_max/PSDz_max[0]-1, label = 'z central frequency')
ax1.plot(time2, LP/LP[0]-1, label = 'Laser power')
ax1.legend()
ax2.plot(time, T, '--', color = 'tab:purple', label = 'Temperature')
ax2.legend()
```

```python
fig = plt.figure( figsize = (14, 7))
gs = fig.add_gridspec(3, hspace=0)
ax1, ax2, ax3 = gs.subplots(sharex=True)
#fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex = True, figsize = (14, 7))
ax1.plot(time2/3600, PSDy_max/1000, label = 'Measured')
ax1.plot(time2/3600, PSDy_max[0]*np.sqrt(LP/LP[0])/1000, label = 'Estimated from power fluctuations')
ax1.legend()
ax2.plot(time2/3600, PSDx_max/1000, label = 'Measured')
ax2.plot(time2/3600, PSDx_max[0]*np.sqrt(LP/LP[0])/1000, label = 'Estimated from power fluctuations')
ax2.legend()
ax2.set_ylabel('Central Frequency (kHz)')
ax3.plot(time2/3600, PSDz_max/1000, label = 'Measured')
ax3.plot(time2/3600, PSDz_max[0]*np.sqrt(LP/LP[0])/1000, label = 'Estimated from power fluctuations')
ax3.legend()
ax3.set_xlabel('Time (hrs)')
fig.tight_layout()

plt.figure()
plt.plot(time2/3600, PSDy_max-PSDy_max[0]*np.sqrt(LP/LP[0]), label = 'y')
plt.plot(time2/3600, PSDx_max-PSDx_max[0]*np.sqrt(LP/LP[0]), label = 'x')
plt.plot(time2/3600, PSDz_max-PSDz_max[0]*np.sqrt(LP/LP[0]), label = 'z')
plt.xlabel('Central Frequency (Hz)')
plt.ylabel('Time (hrs)')
plt.legend()
```
