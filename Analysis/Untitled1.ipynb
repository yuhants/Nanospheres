```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import periodogram
from scipy.fft import fft, ifft
```

```python
N = 10000
fs = 10

t = np.linspace(0, N/fs, N)
x = np.sin(2 * np.pi * 100 * t)
```

```python
xdft = fft(x)
xdft = xdft[0 : int(N/2)]
psdx = (1/(fs*N)) * np.abs(xdft**2)
psdx[1:-1] = 2 * psdx[1:-1]

freq = np.arange(0, fs/N, N/2)
```

```python
f, Pxx_den = periodogram(x, fs)
```

```python
plt.plot(psdx)
plt.plot(Pxx_den)
```
