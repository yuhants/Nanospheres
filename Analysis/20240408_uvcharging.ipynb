```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import butter, sosfilt, stft, welch
from scipy.optimize import curve_fit
import scipy.io as sio
```

```python
params = {'legend.fontsize': 14,
          'axes.labelsize': 18,
          'axes.titlesize': 18,
          'xtick.labelsize': 14,
          'ytick.labelsize': 14}
plt.rcParams.update(params)
```

```python
def low_pass_filtered(timestream, f_lp, fs):
    sos_lp = butter(4, f_lp, 'lp', fs=fs, output='sos')
    filtered = sosfilt(sos_lp, timestream)

    return filtered
```
