```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import butter, sosfilt, stft, welch
from scipy.optimize import curve_fit
import scipy.io as sio
```

## Determine the charge state of the nanosphere

```python
fname = r"C:\Users\yuhan\Documents\Data\charging\20240408_uvcharging.mat"
data = sio.loadmat(fname)
```

```python
data.keys()
```

```python
tt = np.arange(data['Length'][0,0]*data['Tinterval'][0,0], step=data['Tinterval'][0,0])

fig, ax = plt.subplots(1, 1, figsize = (20, 4))
ax_twin = ax.twinx()

ax.plot(tt, data['E'][:,0], 'r.', markersize=12)
ax_twin.plot(tt, data['F'][:,0], 'cyan')

ax.set_ylabel('Demod amp (V)', fontsize=18)
ax.set_xlabel('Time (s)', fontsize=18)
ax_twin.set_ylabel('Phase (rad)', fontsize=18)
ax_twin.tick_params(axis='y', color='k')

fig.tight_layout()
plt.show()
```
