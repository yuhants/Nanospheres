```python
PSDs = []
PSDs2 = []
for i in range(1, 4):
    filename = 'D:\\Experiment\\Cooling\\230516\\Pump down 3\\tt'+str(i)+'.mat'
    data = sio.loadmat(filename)
    tt = np.transpose(data['A'])[0]
    fs = 1/data['Tinterval'][0][0]
    PSD = scisig.welch(tt, fs, nperseg = 10**4)
    PSDs.append(PSD[1])
    tt = np.transpose(data['B'])[0]
    fs = 1/data['Tinterval'][0][0]
    PSD = scisig.welch(tt, fs, nperseg = 10**4)
    PSDs2.append(PSD[1])
freq = PSD[0]
```

```python
plt.figure()
plt.grid()
lb = int(0.02*10000)
ub = int(0.02*33000)
for n, i in enumerate(PSDs):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()

Tz = []
for i in PSDs:
    area = np.trapz(i[lb:ub], freq[lb:ub]) -10**(-9)*(ub-lb)*100
    Tz.append(area)
```

```python
plt.figure()
plt.grid()
lb = int(0.01*88500)
ub = int(0.01*110000)
for n, i in enumerate(PSDs2):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()

Ty = []
for i in PSDs2:
    area = np.trapz(i[lb:ub], freq[lb:ub]) -0.4*10**(-9)*(ub-lb)*100
    Ty.append(area)
```

```python
plt.figure()
plt.grid()
lb = int(0.01*70000)
ub = int(0.01*100000)
for n, i in enumerate(PSDs):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()

Tx = []
for i in PSDs:
    area = np.trapz(i[lb:ub], freq[lb:ub]) -0.4*10**(-9)*(ub-lb)*100
    Tx.append(area)
```

```python
Pressure = [0.89, 0.5, 0.2]
plt.figure()
plt.grid()
plt.grid(which = 'minor', alpha = 0.3)
plt.plot(Pressure, Tx[1:]/Tx[0]*293, 'o', label ='x')
plt.plot(Pressure, Ty[1:]/Ty[0]*293, 'o', label ='y')
plt.plot(Pressure, Tz[1:]/Tz[0]*293, 'o', label ='z')
plt.plot([Pressure[0], Pressure[-1]], [293, 293], '--')
plt.ylabel('Tempertaure (K)')
plt.xlabel('Pressure (mbar)')
plt.yscale('log')
plt.xscale('log')
plt.legend()
```
