```python
import numpy as np
import matplotlib.pyplot as plt

from analysis_utils import *
```

```python
data_file = r"D:\Experiment\Trapping\231010\new_setup_1mbar.csv"
data = np.genfromtxt(data_file, delimiter=',', skip_header=3).T

ffy, ppy = get_psd(data)
ffx, ppx = get_psd(data, channel=2)
ffz, ppz = get_psd(data, channel=3)
```

```python
fig, ax = plt.subplots(figsize = (10, 5))
ax.plot(ffy, ppy, 'b', label='Y detection', alpha=0.7)
ax.plot(ffx, ppx, 'r', label='X detection', alpha=0.7)
ax.plot(ffz, ppz, 'c', label='Z detection', alpha=0.6).033
+
ax.set_title('Pressure = 1 mbar', fontsize=14)
ax.set_xlabel('Frequency (Hz)', fontsize=14)
ax.set_ylabel('Spectral density (V$^2$/Hz)', fontsize=14)

ax.set_yscale('log')
ax.legend(frameon=False, fontsize=14)
ax.tick_params(axis='both', which='both', direction='in', labelsize=14)
```
