```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import scipy.optimize as opt
import scipy.signal as scisig
```

```python
PSDs = []
for i in range(23):
    filename = 'C:\\Users\\Tom\\Documents\\Yale\\data\\Nanosphere\\230501\\Cooling x\\Phase\\tt'+str(i)+'.mat'
    data = sio.loadmat(filename)
    tt = np.transpose(data['F'])[0]
    fs = 1/data['Tinterval'][0][0]
    PSD = scisig.welch(tt, fs, nperseg = 10**4)
    PSDs.append(PSD[1])
freq = PSD[0]
```

```python
plt.figure()
plt.grid()
lb = int(0.01*50000)
ub = int(0.01*120000)
for n, i in enumerate(PSDs):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()
```

```python
T = []
for i in PSDs:
    area = np.trapz(i[lb:ub], freq[lb:ub])
    T.append(area)
```

```python
phase = [90, 95, 100, 105, 110, 120, 130, 140, 85, 80, 75, 70, 65, 60, 55, 50, 40, 30, 20, 10, 0, -10]
plt.plot(phase, T[1:]/T[0]*293, 'o')
plt.ylabel('Tempertaure (K)')
plt.xlabel('Phase (deg)')
```

```python
PSDs = []
PSDs2 = []
for i in range(4):
    filename = 'C:\\Users\\Tom\\Documents\\Yale\\data\\Nanosphere\\230501\\Pump down\\tt'+str(i)+'.mat'
    data = sio.loadmat(filename)
    tt = np.transpose(data['F'])[0]
    fs = 1/data['Tinterval'][0][0]
    PSD = scisig.welch(tt, fs, nperseg = 10**4)
    PSDs.append(PSD[1])
    tt = np.transpose(data['G'])[0]
    fs = 1/data['Tinterval'][0][0]
    PSD = scisig.welch(tt, fs, nperseg = 10**4)
    PSDs2.append(PSD[1])
freq = PSD[0]
```

```python
plt.figure()
plt.grid()
lb = int(0.01*10000)
ub = int(0.01*30000)
for n, i in enumerate(PSDs2):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()

Tz = []
for i in PSDs2:
    area = np.trapz(i[lb:ub], freq[lb:ub]) -10**(-9)*(ub-lb)*100
    Tz.append(area)
```

```python
plt.figure()
plt.grid()
lb = int(0.01*80000)
ub = int(0.01*110000)
for n, i in enumerate(PSDs2):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()

Ty = []
for i in PSDs2:
    area = np.trapz(i[lb:ub], freq[lb:ub]) -0.4*10**(-9)*(ub-lb)*100
    Ty.append(area)
```

```python
plt.figure()
plt.grid()
lb = int(0.01*70000)
ub = int(0.01*100000)
for n, i in enumerate(PSDs):
    plt.plot(freq[lb:ub], i[lb:ub], label = str(n))
plt.yscale('log')
plt.ylabel('PSD (V$^2$/Hz)')
plt.xlabel('Frequency (Hz)')
plt.legend()

Tx = []
for i in PSDs:
    area = np.trapz(i[lb:ub], freq[lb:ub]) -0.4*10**(-9)*(ub-lb)*100
    Tx.append(area)
```

```python
Pressure = [0.6, 0.14, 10**(-4)]
plt.figure()
plt.grid()
plt.grid(which = 'minor', alpha = 0.3)
plt.plot(Pressure, Tx[1:]/Tx[0]*293, 'o', label ='x')
plt.plot(Pressure, Ty[1:]/Ty[0]*293, 'o', label ='y')
plt.plot(Pressure, Tz[1:]/Tz[0]*293, 'o', label ='z')
plt.plot(Pressure, [293, 293, 293], '--')
plt.ylabel('Tempertaure (K)')
plt.xlabel('Pressure (mbar)')
plt.yscale('log')
plt.xscale('log')
plt.legend()
```
