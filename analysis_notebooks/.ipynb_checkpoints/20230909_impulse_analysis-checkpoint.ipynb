```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.signal import welch, stft, butter, lfilter, sosfilt
from scipy.optimize import curve_fit
```

```python
def plot_tod(data, title):
    fig, ax = plt.subplots(figsize = (20, 6))

    ax.plot(data[0], data[1], label="Y detection")
    ax.plot(data[0], data[2], label="X detection")
    ax.plot(data[0], data[3], label="Applied impulse signal")
    try:
        ax.plot(data[0], data[4], label="Feedback signal")
    except:
        pass

    ax.legend(frameon=False, fontsize=18)
    ax.set_xlabel('Time (ms)', fontsize=18)
    ax.set_ylabel('Signal (V)', fontsize=18)
    ax.set_title(title, fontsize=18)
    
    return fig, ax
```

```python
def get_filtered_signal(data, f_hp=20000, f_lp=100000):
    fs = int(np.ceil(1 / (data[0, 1] - data[0, 0]))) * 1000 # Sampling rate in Hz
    
#     sos_hp = butter(8, f_hp, 'hp', fs=fs, output='sos')
#     sos_lp = butter(8, f_lp, 'lp', fs=fs, output='sos')
#     filtered = sosfilt(sos_hp, data[1])
#     filtered_filtered = sosfilt(sos_lp, filtered)
#     return filtered_filtered

    sos = butter(8, [f_hp, f_lp], 'bandpass', output='sos', fs=fs)
    filtered = sosfilt(sos, data[1])
    return filtered
```

```python
def impulse_model(t, t_imp, amp, omega0, gamma, phi, amp0, phi0):
    beta = gamma / 2
    omega1 = np.sqrt(omega0**2 - beta**2)
    
    impulse = amp * np.exp(-1 * beta * (t - t_imp)) * np.sin(omega1 * (t - t_imp) + phi)
    impulse[t < t_imp] = 0
    
#     ret = impulse + amp0 * np.sin(omega1 * t + phi0)
    ret = impulse
    
    return ret
```

```python
def fit_impulse(tt, filtered_timestream, impulse_model, p0=None):
    popt, pcov = curve_fit(impulse_model, tt, filtered_timestream, p0=p0)
    
    return popt, tt, impulse_model(tt, *popt)    
```
